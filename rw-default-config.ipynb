# Resource Watch: Default Vega Configuration
When rendering Vega widgets, the Resource Watch platform (seemingly all variants & children) first applies a [`config`](https://vega.github.io/vega/docs/config/) element, a part of the visualization grammar which provides default styling. This can be found [on the Vizzuality GitHub](https://github.com/Vizzuality/widget-editor/blob/c3dfedf42596c3fdf0e16b19f3c257ec4336f16c/src/packages/core/src/charts/theme.ts).  

This `config` element can be used within [the Vega editor](https://vega.github.io/editor/#/custom/vega) in order for the widget design process to better reflect the ultimate presentation of the figure within RW. The Vega editor has a separate pane for configuration; within a single Vega specification, the contents would be encapsulated within a top-level element named `config`.
